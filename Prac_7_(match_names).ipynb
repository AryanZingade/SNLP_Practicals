```python
import pandas as pd
import numpy as np
```

```python
base_names = pd.read_csv("base_names.csv")
variation_names = pd.read_csv("name_variations.csv")
```

```python
base_names_list = base_names['Base_Name'].tolist()
variations_list = variation_names['Variation'].tolist()
```

```python
import difflib
from fuzzywuzzy import fuzz, process
```

```python
def fuzzy_match_names_difflib(variations, base_names_list, threshold=0.65):
    matched_results = []
    for name in variations:

        best_matches = difflib.get_close_matches(name, base_names_list, n=1, cutoff=threshold)
        if best_matches:
            best_match = best_matches[0]

            similarity_score = difflib.SequenceMatcher(None, name, best_match).ratio() * 100
            matched_results.append((name, best_match, similarity_score))
        else:
            matched_results.append((name, None, 0))
    return matched_results

matched_names = fuzzy_match_names_difflib(variations_list, base_names_list)
```

```python
matched_df = pd.DataFrame(matched_names, columns=['Variation', 'Best_Match', 'Similarity_Score'])
matched_df.to_csv('fuzzy_matched_names.csv', index=False)
matched_df
```
